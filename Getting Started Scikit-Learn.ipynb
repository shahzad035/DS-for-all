```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
#represent the height of group of people in meter
heights = [[1.6],[1.65],[1.7],[1.73],[1.8]]

#represent the weight of group of people in KG
weigts = [[60],[65],[72.3],[75],[80]]
```

```python
plt.title('Weights plotted against Height')
plt.xlabel('Height in Meter')
plt.ylabel('Weight in kilogram')

#Draw a plot aginst height and weight

# Convert a MATLAB style color/line style format string to a (*linestyle*, *marker*, *color*) tuple.
#     Example format strings include:
#     * 'k.': black dot
#     * 'ko': black circles, 'bo' : blue circles, 'ro' : red circles
#     * 'rs': red blocks
#     * '.b': blue dots
#     * 'r--': red dashed lines
#     * 'C2--': the third color in the color cycle, dashed lines
    
plt.plot(heights,weigts,'k.')

#axis range for x and y
plt.axis([1.5,1.85,50,90])
plt.grid(True)
```

```python
#Using the LinearRegression Class for Fitting the Model
from sklearn.linear_model import LinearRegression

#Create and fit the model
model = LinearRegression()
model.fit(X=heights, y=weigts)
```

- Observe that the ***heights*** and ***weights*** are both represented as two-dimensional lists. 
This is because the ***fit()*** function requires both the ***X*** and ***y*** arguments to be 
two-dimensional (of type list or ndarray).

```python
# Make prediction Model
# We want to predict the weight for a person that is
# 1.75m tall. Based on the model, the weight is predicted to be 76.04kg.
weight = model.predict([[1.75]])[0][0]
print(round(weight,2))
```

- In Scikit-learn, you typically use the ***fit()*** function to train a model. Once the model is trained, you use the ***predict()*** function to make a prediction.

```python
plt.title('Weights plotted against Height')
plt.xlabel('Height in Meter')
plt.ylabel('Weight in kilogram')   
plt.plot(heights,weigts,'k.')
plt.axis([1.5,1.85,50,90])
plt.grid(True)

#Plotting Liner Regression Line
plt.plot(heights, model.predict(heights),color='r')
```
